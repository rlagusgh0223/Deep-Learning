```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('bird.jpg',cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
import numpy as np
filter = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])

f_img = cv2.filter2D(img, -1, filter)
cv2_imshow(f_img)
```
